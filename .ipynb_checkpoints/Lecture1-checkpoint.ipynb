## Seasonal Time Plots

```r
ggseasonplot(a10, year.labels=TRUE, year.labels.left=TRUE) + ylab("$ million") + ggtitle("Seasonal plot: antidiabetic drug sales")
```

```r
ggseasonplot(a10, polar=TRUE) + ylab("$ million")
```

```r
ggsubseriesplot(a10) + ylab("$ million") + ggtitle("Subseries plot: antidiabetic drug sales")
```

## Lag plots and autocorrelation


```r
beer <- window(ausbeer, start=1992)
gglagplot(beer)
```

```r
ggAcf(beer,plot=FALSE)
```

```r
ggAcf(beer)
```

```r
elec2 <- window(elec, start=1980)
autoplot(elec2)
```

```r
ggAcf(elec2, lag.max=48)
```

```r
autoplot(goog)
```

```r
ggAcf(goog, lag.max=100)
```

## White noise

```r
wn <- ts(rnorm(36))
autoplot(wn)
```

```r
ggAcf(wn, lag.max=10)
```

```r
ggAcf(wn, lag.max=10,plot=FALSE)
```

```r
hist(wn)
```

```r
meanwn <- mean(wn)
meanwn
```
